```python
# Importing Dependencies
import numpy as np
import random
import matplotlib.pyplot as plt
```

### Linear Regression

```python
# Generate Random Data
n = 1000
X = 10* np.random.random(n) - 5
Y = 4*X + 3 + np.random.normal(loc = 0, scale = 5, size = n)

plt.figure()
plt.scatter(
    X,
    Y
)
```

```python
def Y_based_on_X(x,a,b):
  return a*x +b

# Gradient Descent
a,b = 0,0
l = 0.01
Loss=[]

iter = 50

for i in range(iter):
  Yp=Y_based_on_X(X,a,b)
  loss = np.sum((Yp - Y)**2)/Y.shape[0]
  Loss.append(loss)
  grad_a= np.sum(2*(Yp-Y)*X)/Y.shape[0]
  grad_b= np.sum(2*(Yp-Y))/Y.shape[0]
  a = a - l*grad_a
  b= b- l*grad_b



plt.figure()
plt.scatter(
    X,
    Y
)

plt.scatter(X,Y_based_on_X(X,a,b),c = "red")
```

```python
plt.plot(range(1,iter+1),Loss)
```

```python
# Closed form SOlution
X_0 = []
for i in range(len(X)):
  X_0.append((1,X[i]))
X = np.matrix(X_0)
Y = np.matrix(Y)
beta = (X.T*X).I*X.T*Y.T

beta
```

As you can see the values that we get for the model are extremely close to the 3 and 4 parameters that we initially set up.
