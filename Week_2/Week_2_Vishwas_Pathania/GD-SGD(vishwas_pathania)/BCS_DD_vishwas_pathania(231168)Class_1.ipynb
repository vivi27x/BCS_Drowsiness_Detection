```python
# Importing Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

### Linear Regression

```python
# Generate Random Data
n = 1000
X = 10* np.random.random(n) - 5
Y = 4*X + 3 + np.random.normal(loc = 0, scale = 5, size = n)

plt.figure()
plt.scatter(
    X,
    Y
)
```

```python
def Y_based_on_X(x,a,b):
  return a*x +b

# Gradient Descent
w,b = 0,0
l = 0.01
def descend(x,y,w,b,learning_rate):
  dldw=0
  dldb=0
  N=x.shape[0]
  #loss function=(y-(wx+b))^2
  for xi,yi in zip(x,y):
    dldw += -2*xi*(yi-(w*xi+b))
    dldb += -2*(yi-(w*xi+b))

  w=w-learning_rate*(1/N)*dldw
  b=b-learning_rate*(1/N)*dldb
  return w,b

for epoch in range(n):
  w,b=descend(X,Y,w,b,l)

print(w,b)


plt.figure()
plt.scatter(
    X,
    Y
)

plt.scatter(X,Y_based_on_X(X,w,b),c = "red")
plt.show()
```

### in general increasing n value is increasing the accuracy and making the predictions more stable as the system has more number of points to find the relation also as the accuracy also depends on the distribution of the datapoints so it can also led to decrease in accuracy of one of the parameter

```python
# stochastic gradient descend
# To convert the provided code to use stochastic gradient descent (SGD),
#  we need to modify the descend function to update the parameters  w and  
#  b based on the gradients computed for each individual data point rather than
#  averaging gradients over all data points
def Y_based_on_X(x,a,b):
  return a*x +b

w,b = 0,0
l = 0.01
m=100
def sdescend(x,y,w,b,learning_rate):
  N=x.shape[0]
  #loss function=(y-(wx+b))^2
  for i in range(n):
    xi,yi=x[i],y[i]
    dldw = -2*xi*(yi-(w*xi+b))
    dldb = -2*(yi-(w*xi+b))
    w=w-learning_rate*dldw
    b=b-learning_rate*dldb
  return w,b

for epoch in range(m):
  w,b=sdescend(X,Y,w,b,l)

print(w,b)

plt.figure()
plt.scatter(
    X,
    Y
)

plt.scatter(X,Y_based_on_X(X,w,b),c = "red")
plt.show()
```

As you can see the values that we get for the model are extremely close to the 3 and 4 parameters that we initially set up.
